# TP CSS
*(D'après [Lycée des Flandres](http://www.lyceedesflandres.fr/))*

## Objectif

**Réaliser le style d’une page web**. Vous disposez de trois fichiers :

* `margaret_hamilton.html` est la page web qu’il faut mettre en forme. Il n’y a rien à faire dans ce document, examinez en détail les balises et classes. (L'utilisation de l'outil d'inspecteur de Firefox est très utile...)


* `style.css` est le style de la page web en question. Ce fichier contient des selecteurs vides. Votre travail est de les remplir avec les propriétés et attributs qui conviennent.


* `Margaret_Hamilton.png` est l’objectif visuel final. Vous devez vous approcher le plus possible de ce résultat. *(Ce fichier est une copie d'écran issue directement du corrigé. Comme le site est assez long, il faut zoomer pour voir les détails, quitte à avoir un peu de "flou")*

   
> Tiré d’un magasine publié en juillet 2019, le texte décrit le parcours surprenant de Margaret Hamilton, une informaticienne de la NASA.


# Quelques précisions techniques

* Vous pouvez compléter le fichier `style.css` au choix avec ATOM ou avec les outils de développement web de Firefox (touches de raccourci `CTRL` + `MAJ` + `I`). L'avantage des outils de Firefox est qu'ils permettent de visualiser directement les modifications de style et d'inspecter le document HTML avec la souris. Attention : Ne pas oublier de sauvegarder régulièrement le travail : Firefox ne demande pas de sauvegarder au moment de quitter la page !!


* La police utilisée pour tout le document est `sans-serif`


* Mon fichier `style.css` corrigé fait moins de 60 lignes...


* Afin de vous éviter une galère trop conséquente, vous trouverez ici la liste de toutes les propriétés CSS utilisées dans la correction, par ordre alphabétique :
    * border-bottom-color
    * border-bottom-style
    * color
    * display
    * font-family
    * font-size
    * font-style
    * line-height
    * margin-left
    * margin-right
    * padding-bottom
    * padding-top
    * text-align
    * text-decoration
    * width
Certaines sont employées plusieurs fois, selon l’élément ou la classe, avec des attributs différents.


* Il peut y avoir une légère différence entre le rendu sous Firefox et sous Chrome. (Si possible privilégiez Firefox) Même si vous devez essayer de vous approcher le plus possible du résultat souhaité, il est illusoire d'avoir un résultat "au pixel près".


* Sources utiles :
    * Sur [Mozilla Development Network ou MDN](https://developer.mozilla.org/fr/) : Utiliser leur moteur de recherche pour trouver de l'information sur une propriété CSS. Regardez et testez les exemples.
    * [W3 Schools](https://www.w3.org/) : même chose
    * premier résultat pour [css centrer un élément](https://www.w3.org/Style/Examples/007/center.fr.html)


# Méthodologie

1. **Avant de commencer à écrire du code CSS**, aller sur le site du [Mozilla Development Network ou MDN](https://developer.mozilla.org/fr/) afin de comprendre à quoi correspondent les propriétés ci-dessus


2. En inspectant le document HTML, essayer d'anticiper sur quel élément appliquer telle ou telle propriété CSS


3. Il s'agit ici de se mettre dans la peau d'un [intégrateur web](https://fr.wikipedia.org/wiki/Int%C3%A9grateur_web) : il faut donc appliquer au mieux le design demandé par le [graphiste](https://fr.wikipedia.org/wiki/Graphiste), surtout pas d'apporter votre touche personnelle !! Chaque détail compte : alignement, type de police, gras, souligné, italique, centrage, taille etc.
